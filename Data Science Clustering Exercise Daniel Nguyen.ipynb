```python
%matplotlib inline 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
plt.rcParams['figure.figsize'] = (16,9)
```

```python
df = pd.read_csv (r'C:\Users\Daniel\Desktop\Data Science\edlich-kmeans-A0.csv')
df.head()
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(df.V1, df.V2,df.V3)
```

```python
#A) What is the best k?
# Answer: It looks like that the best k is 5 
```

```python
#B) Can you print the clustering vector?
kmeans = KMeans(n_clusters=5)
kmeans = kmeans.fit(df)
labels = kmeans.predict(df)
C = kmeans.cluster_centers_
labels
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(df.V1,df.V2,df.V3, c = labels)
ax.scatter(C[:,0],C[:,1],C[:,2], marker = '*', c='#050505', s= 1000)
```
